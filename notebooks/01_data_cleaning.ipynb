```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

```python
#=================================
#       URL e Fileve te CSV
#=================================


RAW_TABLE = "../data/raw/laliga_table/"
RAW_SHT = "../data/raw/shooting/"
RAW_GOAL = "../data/raw/goalkeeping/"
RAW_SS = "../data/raw/squad_standart_stats/"

PROCESSED = "../data/processed/"
```

```python
#===============================
#     Bashkimi i Fileve te CSV
#===============================


def lexo_folder(folder_path):
    te_gjitha = []

    for skedari in sorted(glob.glob(folder_path + "*.csv")):
        emri = os.path.basename(skedari)
        sezoni = ''.join(filter(lambda x: x.isdigit() or x == "-", emri))
        sezoni = sezoni.strip('-')


        df = pd.read_csv(skedari, encoding="utf-8", encoding_errors="replace")
        df["sezoni"]=sezoni

        te_gjitha.append(df)


    rezultati = pd.concat(te_gjitha, ignore_index=True)
    return rezultati
```

```python
#===============================
#   Lexo te gjitha filet t CSV
#===============================



df_tabel = lexo_folder(RAW_TABLE)
df_sht = lexo_folder(RAW_SHT)
df_goal = lexo_folder(RAW_GOAL)
df_ss = lexo_folder(RAW_SS)


print(df_tabel.shape)
print(df_sht.shape)
print(df_goal.shape)
print(df_ss.shape)
```

```python
# ============================================================
# PASTRO LALIGA TABLE
# ============================================================

hiq_kolonat = ["Rk", "Attendance", "Top Team Scorer", "Goalkeeper", "Notes"]
df_tabel = df_tabel.drop(columns=hiq_kolonat)

print(df_tabel.head(3))
print()
print("Kolonat:", df_tabel.columns.tolist())
print("Dimensionet:", df_tabel.shape)
```

```python
# ============================================================
# PASTRO EMRAT E EKIPEVE
# ============================================================

# Hiq çdo karakter që nuk është shkronjë, numër ose hapësirë
df_tabel["Squad"] = df_tabel["Squad"].str.replace(r'[^\w\s]', '', regex=True)
df_tabel["Squad"] = df_tabel["Squad"].str.strip()

# ============================================================
# STANDARDIZO EMRAT E EKIPEVE
# ============================================================

emrat = {
    "Atltico Madrid"  : "Atletico Madrid",
    "Mlaga"           : "Malaga",
    "Alavs"           : "Alaves",
    "Dep La Corua"    : "Deportivo La Coruna",
    "Racing Sant"     : "Racing Santander",
    "Sporting Gijn"   : "Sporting Gijon",
    "Legans"          : "Leganes",
    "Cdiz"            : "Cadiz",
    "Almera"          : "Almeria",
    "Crdoba"          : "Cordoba",
    "Hrcules"         : "Hercules",
    "Gimnstic"        : "Gimnastic",
}

df_tabel["Squad"] = df_tabel["Squad"].replace(emrat)


# Kontrollo rezultatin
print(df_tabel["Squad"].unique())
```

```python
# ============================================================
# PASTRO SHOOTING
# ============================================================

# Hiq rreshtat bosh
df_sht = df_sht.dropna(subset=["Squad"])

# Hiq kolonat e panevojshme
df_sht = df_sht.drop(columns=["# Pl", "90s"])

# Pastro emrat e ekipeve
df_sht["Squad"] = df_sht["Squad"].str.replace(r'[^\w\s]', '', regex=True)
df_sht["Squad"] = df_sht["Squad"].str.strip()

# Standardizo emrat
df_sht["Squad"] = df_sht["Squad"].replace(emrat)

# Kontrollo rezultatin
print(df_sht.head(3))
print()
print("Vlerat NaN:")
print(df_sht.isnull().sum())
print()
print("Dimensionet:", df_sht.shape)
```

```python
# ============================================================
# PASTRO GOALKEEPING
# ============================================================

# Hiq kolonat e panevojshme
kolonat_hiq = ["# Pl", "Starts", "Min", "90s", "W", "D", "L", "Save%.1"]
df_goal = df_goal.drop(columns=kolonat_hiq)

# Pastro emrat e ekipeve
df_goal["Squad"] = df_goal["Squad"].str.replace(r'[^\w\s]', '', regex=True)
df_goal["Squad"] = df_goal["Squad"].str.strip()
df_goal["Squad"] = df_goal["Squad"].replace(emrat)

# Kontrollo rezultatin
print(df_goal.head(3))
print()
print(df_goal.dtypes)
print()
print("Vlerat NaN:")
print(df_goal.isnull().sum())
print()
print("Dimensionet:", df_goal.shape)
```

```python
# ============================================================
# BASHKO TE GJITHA NE NJE DATAFRAME
# ============================================================

# Bashko me left join — df_table si baza
df_master = df_tabel.merge(df_sht,  on=["Squad", "sezoni"], how="left")
df_master = df_master.merge(df_goal, on=["Squad", "sezoni"], how="left")
df_master = df_master.merge(df_ss,   on=["Squad", "sezoni"], how="left")

# Kontrollo rezultatin
print(df_master.head(3))
print()
print("Dimensionet:", df_master.shape)
print()
print("Vlerat NaN per kolone:")
print(df_master.isnull().sum())
```

```python
# ============================================================
# EKSPORTO NE PROCESSED
# ============================================================

# Krijo folderin nese nuk ekziston
os.makedirs(PROCESSED, exist_ok=True)

# Eksporto
df_master.to_csv(PROCESSED + "laliga_master.csv", index=False)

print("U eksportua me sukses!")
print("Lokacioni:", PROCESSED + "laliga_master.csv")
print("Dimensionet finale:", df_master.shape)
```
